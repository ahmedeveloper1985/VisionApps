# High Pass Filters are used to enhance or sharpen parts of the image with high frequency (Edges)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def showimg(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

# Edge Detection (Sobel) - Automatic Method

```python
img_orig = cv2.imread("./images/city_hall_zoom.png")
img_t = cv2.cvtColor(img_orig,cv2.COLOR_BGR2GRAY)

sobelX = cv2.Sobel(img_t, cv2.CV_64F, 1, 0)
sobelY = cv2.Sobel(img_t, cv2.CV_64F, 0, 1)

sobelX = cv2.convertScaleAbs(sobelX)
sobelY = cv2.convertScaleAbs(sobelY)

sobelCombined = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)

showimg(img_orig)
showimg(sobelX)
showimg(sobelY)
showimg(sobelCombined)
```

# Edge Detection (Sobel) - Custom Filters Method

```python
img_orig = cv2.imread("./images/city_hall_zoom.png")
img_t = cv2.cvtColor(img_orig,cv2.COLOR_BGR2GRAY)

sobel_x = np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]])

sobel_y = np.array([[-1,-2,-1],
                  [0,0,0],
                  [1,2,1]])

sobelX = cv2.filter2D(img_t,-1,sobel_x)
sobelY = cv2.filter2D(img_t,-1,sobel_y)

sobelCombined = cv2.addWeighted(sobelX, 0.5, sobelY, 0.5, 0)

showimg(img_orig)
showimg(sobelX)
showimg(sobelY)
showimg(sobelCombined)
```

# Canny Edge Detection:

1- Apply Gaussian Blur to filter out noise<br>
2- Apply Soble Edge Detection to detect edges<br>
3- Apply non-maxima suppression to isolate strongest edges<br>
4- Apply hystresis to isolate best edges (Lower/Upper Bounds)

```python
img = cv2.imread("./images/sunflower.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,CMAP = "gray")
plt.show()

#Tight Threshold
canny = cv2.Canny(img,30,100)
plt.imshow(canny,CMAP = "gray",)
plt.show()

#Wide Threshold
canny = cv2.Canny(img,180,240)
plt.imshow(canny,CMAP = "gray")
plt.show()

#Automatic Canny
canny = imutils.auto_canny(img)
plt.imshow(canny,CMAP = "gray")
plt.show()
```
