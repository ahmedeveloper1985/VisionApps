# Recurrent Neural Network

RNN are non-linear function approximators.

They maintain an internal memory for each element called states.

Output does not depend only on input, but also takes into account passed outputs.

# Long Short Term Memory Cells (LSTM)

Was introduces to help solve the vanishing gradient problem, LSTMS Hold on information for many steps further.

# 

# 

# 

# Learn Gate: Combine The short term memory and the event and ignore a part of the result

# 

# Forget Gate: Takes the long term memory and decide what to keep and what to forget

# 

# Remember Gate: Combine The outputs of the forget gate and learn gate together

# 

# 

# Use Gate: We take what's useful from the long term memory, and what's useful from the short term memory and create a new long term memory
