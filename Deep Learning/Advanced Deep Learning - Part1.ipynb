# Region Convolutional Neural Network (R-CNN)

1- Segment the image into multiple regions of interest (ROIs), this process is called region proposal <br/>
2- Pass these regions to an RCNN for identification <br/>
3- Include an extra "background" label to classify unknown regions

# 

# Fast Region Convolutional Neural Network (Fast R-CNN)

1- Image is paseed through the network once<br/>
2- ROI Pooling is used to segment ROI in the image without cropping and having to pass multiple image regions through the whole network as with the original R-CNN<br/>

Problem: None object regions are included in the analysis and passed to the final layers for classification

# 

# Faster Region Convolutional Neural Network (Faster R-CNN)

Same as last network with the exception of introduction of a new Region Proposal Network (RPN) that will filter out none object regions
