# Low Pass Filters are used to block some high frequence content in the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def showimg(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
img = cv2.imread("./images/noisy1.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img,cmap = "gray")
plt.show()
```

```python
blurred = cv2.GaussianBlur(img,(5,5),0)

plt.imshow(blurred,cmap = "gray")
plt.show()
```
