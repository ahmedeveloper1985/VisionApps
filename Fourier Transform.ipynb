```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Fourier Transform is a mean to measure the the frequency of multiple regions within an image.

# Frequency is high in image areas with high degrees of pixel intensity changes.

```python
def ft_image(image):
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image = image/255
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    frequency_tx = 20*np.log(np.abs(fshift))   
    return frequency_tx
```

```python
img = cv2.imread("./images/high_freq.jpg")

ft = ft_image(img)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.show()
plt.imshow(ft,cmap = "gray")
plt.show()
```
