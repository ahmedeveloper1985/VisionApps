```python
import cv2
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

# Hough Lines

```python
img = cv2.imread("./images/phone.jpg")
orig = img.copy()
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

canny = imutils.auto_canny(img)

plt.imshow(img,cmap="gray")
plt.show()
plt.imshow(canny,cmap="gray")
plt.show()
```

```python
rho = 1
theta = np.pi/180
threshold = 60
min_line_length = 100
max_line_gap = 5


lines = cv2.HoughLinesP(canny,rho,theta,threshold,minLineLength = min_line_length,maxLineGap = max_line_gap)


output = orig.copy()
for [(x1,y1,x2,y2)] in lines:
    cv2.line(output,(x1,y1),(x2,y2),(255,255,0),2)
    
cv2.imshow("Image",output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Hough Circles

```python
img = cv2.imread("./images/round_farms.jpg")
orig = img.copy()
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(img, (3, 3), 0)

canny = imutils.auto_canny(img)

plt.imshow(img,cmap="gray")
plt.show()
plt.imshow(canny,cmap="gray")
plt.show()
```
